```python
new_df = pd.DataFrame(columns=['Tag_Number', 'SciName', 'Site'])
```

```python
for i in set(df['Date']):
    new_df[i] = np.nan
```

```python
new_df[:5]
```

```python
for i in df.index:
    new_row = []
    new_row.append(df['Tag_Number'][i])
    new_row.append(df['SciName'][i])
    new_row.append(df['Site'][i])
    for j in range(4):
        new_row.append(0)
    date = df['Date'][i]
    new_df.loc[i] = new_row
    new_df[date][i] = 1
```
